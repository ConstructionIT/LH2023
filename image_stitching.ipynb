```python
from imutils import paths
import numpy as np
import imutils
import cv2
```

```python
imagePaths = sorted(list(paths.list_images("/home/enfycius/Desktop/ImgSets/")))
```

```python
images = []
```

```python
for imagePath in imagePaths:
    image = cv2.imread(imagePath)
    images.append(image)
```

```python
len(images)
```

```python
stitcher = cv2.createStitcher() if imutils.is_cv3() else cv2.Stitcher_create()
```

```python
(status, stitched) = stitcher.stitch(images)
```

```python
plt.imshow(stitched)
plt.show()
```
